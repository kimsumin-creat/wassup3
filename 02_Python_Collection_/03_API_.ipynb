```python
import requests, json, pandas as pd
```

```python
URL='http://www.krei.re.kr:18181/chart/main_chart/index/kind/W/sdate/2019-01-01/edate/2019-12-31'
```

```python
resp = requests.get(URL)
resp
```

```python
data1 = resp.text
print(type(data1), len(data1))
print(data1)
```

```python
data2=resp.json()
print(type(data2), len(data2))
print(data2)
```

```python
data2
```

```python
## data2에서 id와 data, symbol,open,close추출하기
```

```python
id_=[]
date=[]
symbol=[]
open_=[]
close=[]
for item in data2:
    date.append(item['date'])
    symbol.append(item['symbol'])
    open_.append(item['open'])
    close.append(item['close'])
    
```

```python
df = pd.DataFrame({
    'id' : id_, #칼럼명:데이터
    'date' : date,
    'symbol' :symbol,
    'open' : open_,
    'close' : close,
})
```

## 퀴즈: 거래금액,건축년도, 년, 법정동, 아파트, 월, 일, 전용면적, 지번, 지역코드, 층데이터 추출

```python
# 반복되는 코드 정리
item_list = ['거래금액', '건축년도', '년', '법정동', '아파트', '월', '일', '전용면적', '지번', '지역코드', '층']

for item in items:
    for tag in item_list:
        print(item.select_one(tag).text.strip(), end=' ')
    print()
```
